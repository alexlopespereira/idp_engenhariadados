### Carregando dados com a biblioteca sqlite3

```python
# %sql sqlite:////content//world.sqlite

import sqlite3

conn = sqlite3.connect('//content//world.sqlite')

cursor = conn.cursor()

query = "select * from world limit 3"

cursor.execute(query)

rows = cursor.fetchall()
display(rows)

conn.close()
```

### Carregando dados com o pandas

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('//content//world.sqlite')

sql_query = "select * from world limit 3"

df = pd.read_sql_query(sql_query, conn)

df.head()
```

```python
# Exemplo de Query em multiplas linhas
sql_query = "SELECT Country, GDP2017 FROM world LIMIT 3"

df = pd.read_sql_query(sql_query, conn)
df.head()
```
