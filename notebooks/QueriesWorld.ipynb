*   Calcular a média e a soma da população de todos os registros da tabela world.
*   Calcular a média e a soma da população de todos os registros da tabela world do continente de Africano.
*   Retornar a máxima população do continente do Europeu.
*   Calcular a média da população dos continentes cujo nome começa com a letra A.
*   Retornar os nomes dos continentes, sem repetições.
*   Contar a quantidade de continentes.


```python
import sqlite3
import pandas as pd
# Substitua 'nome_do_arquivo.sqlite' pelo nome real do seu arquivo
con = sqlite3.connect('world.sqlite')
cursor = con.cursor()
```

```python
cursor.execute("select * from world;")
nomes_colunas = [descricao[0] for descricao in cursor.description]
df = pd.DataFrame(cursor.fetchall(), columns=nomes_colunas)
display(df)
```

```python
cursor.execute("select avg(Population2017), sum(Population2017) from world;")
nomes_colunas = [descricao[0] for descricao in cursor.description]
df = pd.DataFrame(cursor.fetchall(), columns=nomes_colunas)
display(df)
```

```python
cursor.execute("select avg(Population2017), sum(Population2017) from world where Continent='Africa';")
nomes_colunas = [descricao[0] for descricao in cursor.description]
df = pd.DataFrame(cursor.fetchall(), columns=nomes_colunas)
display(df)
```

```python
cursor.execute("select max(Population2017) from world where Continent='Europe';")
nomes_colunas = [descricao[0] for descricao in cursor.description]
df = pd.DataFrame(cursor.fetchall(), columns=nomes_colunas)
display(df)
```

```python
cursor.execute("select sum(Population2017) from world where Continent like 'A%';")
nomes_colunas = [descricao[0] for descricao in cursor.description]
df = pd.DataFrame(cursor.fetchall(), columns=nomes_colunas)
display(df)
```
